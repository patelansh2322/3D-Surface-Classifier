```python
import pandas as pd
import numpy as np
import sympy as sp
import random
import math
```

```python
df = pd.DataFrame(columns=['Equation', 'a', 'b', 'c', 'fx', 'fy', 'fxx', 'fyy', 'fxy', 'D', 'Result'])
```

```python
x, y, a, b, c = sp.symbols('x y a b c')
```

```python
for i in range(75):
    randA = random.randint(-50, 50)
    randB = random.randint(-50, 50)
    f = a*x**2 + b*y**2
    fx = sp.diff(f, x).subs({a:randA, b:randB})
    fy = sp.diff(f, y).subs({a:randA, b:randB})
    fxx = sp.diff(fx, x).subs({a:randA, b:randB})
    fyy = sp.diff(fy, y).subs({a:randA, b:randB})
    fxy = sp.diff(fx, y).subs({a:randA, b:randB})
    D = float(fxx*fyy-fxy**2)
    if D==0:
        result = "Inconclusive"
    elif D<0:
        result = "Saddle Point"
    elif D>0 and fxx>0:
        result = "Minimum"
    elif D>0 and fxx<0:
        result = "Maximum"
    else:
        result = "Inconclusive"
    df.loc[len(df)] = [f, randA, randB, 0, fx, fy, fxx, fyy, fxy, D, result]  
```
